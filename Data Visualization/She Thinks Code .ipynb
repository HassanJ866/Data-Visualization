```python
df3 = pd.read_csv('DATA/countries.csv')
df3.head()
```

```python
US_pop = df3[df3.country == "United States"]
China_pop = df3[df3.country == "China"]

plt.plot(US_pop.year,US_pop.population/10 ** 6)
plt.plot(China_pop.year,China_pop.population/10 ** 6)
plt.title("US Population vs China Population")
plt.xlabel("Years")
plt.ylabel("Population")
plt.legend(['United States', 'China'])
plt.show()
```

```python
us_per = (US_pop.population/US_pop.population.iloc[0] * 100)
china_per = (China_pop.population/China_pop.population.iloc[0] * 100)

plt.plot(US_pop.year,us_per, '*')
plt.plot(China_pop.year,china_per, '--')
plt.show()
```
